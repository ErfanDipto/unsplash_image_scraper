```python
def get_img_urls(query=None, results_per_page=20, page_num=1):
  url = f"https://unsplash.com/napi/search/photos?query={query}&per_page={result_per_page}&page={page_num}"

  if query == None:
    raise ValueError("Provide the query to search")

  headers = {
    'sec-ch-ua': '"Brave";v="119", "Chromium";v="119", "Not?A_Brand";v="24"',
    'Referer': 'https://unsplash.com/s/photos/water',
    'accept-language': 'en-US',
    'sec-ch-ua-mobile': '?0',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
    'sec-ch-ua-platform': '"Windows"',
    'Cookie': 'require_cookie_consent=false'
  }

  resp = r.request("GET", url, headers=headers)

  if resp.status_code != 200:
    raise ConnectionError("Problem getting response")

  img_result_list = resp.json()["results"]
  raw_img_lists = [img_url["urls"]["raw"] for img_url in img_result_list if img_url["premium"]==False]
  img_urls = [img.split("?")[0] for img in raw_img_lists if "photo" in img]

  return img_urls
```

```python
def download_api_images(img_urls, max_download=5, download_count=0, dir="images",tag=""):

  to_go = max_download - download_count

  headers = {
  'sec-ch-ua': '"Brave";v="119", "Chromium";v="119", "Not?A_Brand";v="24"',
  'Referer': 'https://unsplash.com/s/photos/water',
  'accept-language': 'en-US',
  'sec-ch-ua-mobile': '?0',
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36',
  'sec-ch-ua-platform': '"Windows"',
  'Cookie': 'require_cookie_consent=false'
  }

  for url in img_urls:
    resp = r.get(url, headers=headers)

    file_name = url.split(".com/")[1]

    if not os.path.exists(dir):
      os.makedirs(dir)

    with open(f"{dir}/{tag}_{file_name}.jpeg", "wb") as f:
      f.write(resp.content)

    download_count+=1

    to_go = max_download - download_count

    if to_go==0:
      return download_count

  if to_go>0:
    return download_count

```

```python
def scrape(keyword=None, max_download=5):
  if keyword==None:
    raise ValueError("Provide what image to search")
  page = 1
  img_urls = get_img_urls(query=keyword, page_num=page)
  download_count = download_api_images(img_urls=img_urls, dir=f"{keyword}_images", max_download=max_download, download_count=0,tag=keyword)

  while(download_count<max_download):
    page+=1
    img_urls = get_img_urls(query=keyword, page_num=page)
    download_count = download_api_images(img_urls=img_urls, dir=f"{keyword}_images", max_download=max_download, download_count=download_count,tag=keyword)
```

```python
scrape("sweets", 20)
```
